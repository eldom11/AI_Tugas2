```python
graph = {
    'Arad': ['Zerind', 'Sibiu', 'Timisoara'],
    'Zerind': ['Arad', 'Oradea'],
    'Sibiu': ['Arad', 'Oradea', 'Fagaras', 'Rimnicu Vilcea'],
    'Timisoara': ['Arad', 'Lugoj'],
    'Oradea': ['Zerind', 'Sibiu'],
    'Fagaras': ['Sibiu', 'Bucharest'],
    'Rimnicu Vilcea': ['Sibiu', 'Pitesti', 'Craiova'],
    'Lugoj': ['Timisoara', 'Mehadia'],
    'Pitesti': ['Rimnicu Vilcea', 'Bucharest', 'Craiova'],
    'Mehadia': ['Lugoj', 'Drobeta'],
    'Drobeta': ['Mehadia', 'Craiova'],
    'Craiova': ['Drobeta', 'Rimnicu Vilcea', 'Pitesti'],
    'Bucharest': ['Fagaras', 'Pitesti', 'Giurgiu', 'Urziceni'],
    'Giurgiu': ['Bucharest'],
    'Urziceni': ['Bucharest', 'Hirsova', 'Vaslui'],
    'Hirsova': ['Urziceni', 'Eforie'],
    'Vaslui': ['Urziceni', 'Iasi'],
    'Iasi': ['Vaslui', 'Neamt'],
    'Eforie': ['Hirsova'],
    'Neamt': ['Iasi']
}
```

```python
def bfs(graph, start, goal):
    visited = set()
    queue = [[start]]

    if start == goal:
        return "Start and goal are the same."

    while queue:
        path = queue.pop(0)
        node = path[-1]
```

```python
def dfs(graph, start, goal):
    visited = set()
    stack = [[start]]

    if start == goal:
        return "Start and goal are the same."

    while stack:
        path = stack.pop()
        node = path[-1]

        if node not in visited:
            neighbors = graph[node]
            for neighbor in neighbors:
                new_path = list(path)
                new_path.append(neighbor)
                stack.append(new_path)

                if neighbor == goal:
                    return new_path

            visited.add(node)

    return "No path found."
```

```python
# Rute perjalanan dari Arad ke Bucharest menggunakan BFS
bfs_route = bfs(graph, 'Arad', 'Bucharest')
print("Rute dari Arad ke Bucharest menggunakan BFS:", bfs_route)

# Rute perjalanan dari Arad ke Bucharest menggunakan DFS
dfs_route = dfs(graph, 'Arad', 'Bucharest')
print("Rute dari Arad ke Bucharest menggunakan DFS:", dfs_route)
```
